```python
"""This is my Matrimonal Application System
Author: Vigneshrajan
created on: 14-09-2022
copyrights: None"""

import random
import logging
import csv
#filename = 'vigneshrajan.csv'
file= 'vr.csv'
print('welcome to my Matrimoal System')

def Insertion():
    u=random.randint(101,201)
    uniqueId=u
    print('Unique id:',u)
    name=input('Enter Name:')
    gender=input('Enter Gender:')
    Age = int(input('Enter Age:'))
    Occupation=input('Enter Occupation:')
    salary=int(input('Enter salary:'))
    Likes=input('Enter Likes:')
    Dislikes=input('Enter Dislikes')
    Location=input('Enter Location:')


    with open(file,'a',newline='')as f:
        csvwriter=csv.writer(f)
        csvwriter.writerow([uniqueId,name,gender,Age,Occupation,salary,Likes,Dislikes,Location])
        
def Retrival():
    with open(file,'r') as f:
        print( '*******************************************************************************************************************')
        print('uniqueId\tname\tgender\tAge\tOccupation\tsalary\tLikes\tDislikes\tLocation')
        print('*******************************************************************************************************************')

        data=csv.reader(f)
        for r in list(data):
            print(f'{r[0]}\t\t{r[1]}\t\t{r[2]}\t\t{r[3]}\t\t{r[4]}\t\t{r[5]}\t\t{r[6]}\t\t{r[7]}\t\t{r[8]}')
def Searching():
    s=input("Enter the search data:")
    print( '***********************************************RESULT********************************************************************')
    with open('vr.csv','r') as file:
        cr=csv.reader(file)
        li=list(cr)
        result=[i for i in li if s in i]
        if len(result)>0:
            for j in list(result):
                print(j[0],'\t',j[1],'\t',j[2],'\t',j[3],'\t',j[4],'\t',j[5],'\t',j[6],'\t',j[7],'\t',j[8])
        else:
            print( '************************************RESULT NOT FOUND********************************************************************')
def Deletion():
    s=input('Enter data to delete')
    with open('vr.csv','r') as fi:
        d=csv.reader(fi)
        n=list(d)
        result=[i for i in n if s!=i[0]]
    with open('vr.csv','w',newline="")as fi:
        csvwriter=csv.writer(fi)
        csvwriter.writerows(result)
        if(len(result)==len(n)):
            print('Not available')
        else:
            print('updated list')
    Retrival()

def menu():
    print('Menu')
    print('0. Exit')
    print('1. Insertion')
    print('2. Modification')
    print('3. Retrival')
    print('4. Searching')
    print('5. Recommendation')
    print('6. Deletion')
    choice = eval(input('Enter the choice'))
    return choice
while True:
    choice=menu()
    if choice==0:
        break
    elif choice==1:
        Insertion()
    elif choice==2:
        Modification()
    elif choice==3:
        Retrival()
    elif choice==4:
        Searching()
    elif choice==5:
        Recommendation()
    elif choice==6:
        Deletion()
    else:
        pass

    
    

        
```
